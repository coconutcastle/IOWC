DAI PDSI correlations with PHYDA, LMR, and Reanalysis for Tabora

```python
import pandas as pd
from scipy.stats import pearsonr

# load .csv files containing the annual means
dai_pdsi_avg = pd.read_csv('./tabora_pdsi_avg.csv')
dai_pdsi_sc_avg = pd.read_csv('./tabora_pdsi_sc_avg.csv')
phyda_pdsi_avg = pd.read_csv('../../PHYDA/phyda_tabora_all_pdsi.csv')
lmr_pdsi_prate_avg = pd.read_csv('../../LMR/tabora_files/prate_pdsi_mcrun_mean.csv')
reanalysis_prate_avg = pd.read_csv('../../PHYDA/tabora_reanalysis_redo.csv')
```

```python
def isValidSeries(single_series):
    return (single_series.size == 1 and single_series.iloc[0] != float('nan') and single_series.iloc[0] != -999 and single_series.iloc[0] != -999.9)

print(isValidSeries(dai_pdsi_avg['Average'].loc[dai_pdsi_avg['Year'] == 1894]))
print(isValidSeries(reanalysis_prate_avg['Annual'].loc[reanalysis_prate_avg['Year'] == 2016]))
```

```python
# find years in common
common_years = []

for year in range(1800, 2100):
    valid_counter = 0
    to_test = [dai_pdsi_avg['Average'].loc[dai_pdsi_avg['Year'] == year],
               dai_pdsi_sc_avg['Average'].loc[dai_pdsi_sc_avg['Year'] == year],
               phyda_pdsi_avg['PDSI Mean'].loc[phyda_pdsi_avg['Year'] == year],
               lmr_pdsi_prate_avg['Mean PDSI'].loc[lmr_pdsi_prate_avg['Year'] == year],
               lmr_pdsi_prate_avg['Mean PRATE'].loc[lmr_pdsi_prate_avg['Year'] == year],
               reanalysis_prate_avg['Annual'].loc[reanalysis_prate_avg['Year'] == year]]

    for series in to_test:
        if isValidSeries(series):
            valid_counter+=1

    if valid_counter == len(to_test):
        common_years.append(year)

print(common_years)
```

```python
class Data_to_compare:
    def __init__(self, name, df, array, year_column):
        self.name = name
        self.df = df
        self.array = array
        self.year_column = year_column
```

```python
# reduce data to same size arrays of only the years in common

dai_pdsi_data = Data_to_compare('dai_pdsi', dai_pdsi_avg['Average'], [], dai_pdsi_avg['Year'])
dai_pdsi_sc_data = Data_to_compare('dai_pdsi_sc', dai_pdsi_sc_avg['Average'], [], dai_pdsi_sc_avg['Year'])
phyda_pdsi_data = Data_to_compare('phyda_pdsi', phyda_pdsi_avg['PDSI Mean'], [], phyda_pdsi_avg['Year'])
lmr_pdsi_data = Data_to_compare('lmr_pdsi', lmr_pdsi_prate_avg['Mean PDSI'], [], lmr_pdsi_prate_avg['Year'])
lmr_prate_data = Data_to_compare('lmr_prate', lmr_pdsi_prate_avg['Mean PRATE'], [], lmr_pdsi_prate_avg['Year'])
reanalysis_prate_data = Data_to_compare('reanalysis_prate', reanalysis_prate_avg['Annual'], [], reanalysis_prate_avg['Year'])

all_data = [dai_pdsi_data, dai_pdsi_sc_data, phyda_pdsi_data, lmr_pdsi_data, lmr_prate_data, reanalysis_prate_data]

for year in common_years:
    for data in all_data:
        data.array.append(data.df.loc[data.year_column == year].to_numpy()[0])

for data in all_data:
    print('\n=============================\n')
    print(f'Array size is: {len(data.array)}')
    print(data.array)
```

```python
# correlate the lmr, phyda, and reanalysis data with dai and dai sc, not with eachother (already previously done)

preexisting_data = [phyda_pdsi_data, lmr_pdsi_data, lmr_prate_data, reanalysis_prate_data]

for data in preexisting_data:
    print('\n============ FOR ' + data.name + ' =============\n')
    pdsi_correl = pearsonr(data.array, dai_pdsi_data.array)
    pdsi_sc_correl = pearsonr(data.array, dai_pdsi_sc_data.array)
    print(f'Correlation with DAI PDSI: {pdsi_correl[0]}')
    print(f'Correlation with DAI PDSI SC: {pdsi_sc_correl[0]}')
```

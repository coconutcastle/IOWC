PRATE AND PDSI CORRELATIONS

Note: Using scipy's Pearson correlation coefficient function

```python
from scipy.stats import pearsonr

# With original prate values (not per day or per year)

prate_df = pd.read_csv('tabora_files/tabora_lmr_original_all_mcrun.csv')
pdsi_df = pd.read_csv('tabora_files/tabora_lmr_pdsi_all_mcrun.csv')

print(prate_df)
print("\n=================================================\n")
print(pdsi_df)
```

```python
# Scale and translate the data so that it falls on a range between -1 and 1

from sklearn.preprocessing import MinMaxScaler

prate_array = prate_df.drop(['Year'], 1)        #drop year column
pdsi_array = pdsi_df.drop(['Year'], 1)

prate_array = prate_array.to_numpy()
pdsi_array = pdsi_array.to_numpy()

print(prate_array)
print('\n==========================================\n')
print(pdsi_array)
```

```python
min_max_scaler_prate = MinMaxScaler(feature_range=(-1, 1))
min_max_scaler_pdsi = MinMaxScaler(feature_range=(-1, 1))

min_max_scaler_prate.fit(prate_array)
min_max_scaler_pdsi.fit(pdsi_array)

print(f'PRATE data range is: {min_max_scaler_prate.data_range_}')
print(f'PRATE feature range is: {min_max_scaler_prate.feature_range}')
print('')
print(f'PDSI data range is: {min_max_scaler_pdsi.data_range_}')
print(f'PDSI feature range is: {min_max_scaler_prate.feature_range}')

prate_array_scaled = min_max_scaler_prate.transform(prate_array)
pdsi_array_scaled = min_max_scaler_pdsi.transform(pdsi_array)

print(prate_array_scaled)
print('\n==================================================\n')
print(pdsi_array_scaled)
```

```python
# testing columns

print(prate_array_scaled[:, 1])
print('\n==================================================\n')
print(pdsi_array_scaled[:, 1])

print('\n==================================================\n')
print(prate_array_scaled[:, 0].shape)
print(pdsi_array_scaled[:, 0].shape)
print('')
print(prate_array_scaled.shape)
print(pdsi_array_scaled.shape)
```

```python
corr_value_list = []
mcrun_list = list(range(0, 20))
```

```python
for i in range(20):
    # all_years_index_array_prate = prate_df[f'MCrun Index {i}']
    # all_years_index_array_pdsi = pdsi_df[f'MCrun Index {i}']

    all_years_index_array_prate = prate_array_scaled[:, i]
    all_years_index_array_pdsi = pdsi_array_scaled[:, i]

    corr_coef = pearsonr(all_years_index_array_prate, all_years_index_array_pdsi)
    corr_value_list.append(corr_coef[0])

print(corr_value_list)
```

GET MCRUN MEANS

```python
from scipy.stats import pearsonr

# With original prate values (not per day or per year)

prate_df = pd.read_csv('tabora_files/tabora_lmr_original_all_mcrun.csv')
pdsi_df = pd.read_csv('tabora_files/tabora_lmr_pdsi_all_mcrun.csv')

print(prate_df)
print("\n=================================================\n")
print(pdsi_df)
```

```python
print(prate_df.shape)
```

```python
prate_mcrun_mean_list = []
pdsi_mcrun_mean_list = []

for row in range(prate_df.shape[0]):
    row_mean_prate = 0
    row_mean_pdsi = 0
    for column in range(prate_df.shape[1] - 1):
        row_mean_prate += prate_df.iloc[row][f'MCrun Index {column}']
        row_mean_pdsi += pdsi_df.iloc[row][f'MCrun Index {column}']

    prate_mcrun_mean_list.append((row_mean_prate/(prate_df.shape[1] - 1)))
    pdsi_mcrun_mean_list.append((row_mean_pdsi/(prate_df.shape[1] - 1)))

print(prate_mcrun_mean_list)
print('\n=======================\n')
print(pdsi_mcrun_mean_list)
```

```python
#testing if actually mean

test_pdsi = pdsi_df.iloc[1]
test_pdsi = test_pdsi.to_numpy()
test_pdsi = test_pdsi[1:]
# test_prate = test_prate.to_numpy
print(test_pdsi)
print(type(test_pdsi))
print(test_pdsi.shape)

print(np.mean(test_pdsi))
```

```python
year_list = list(range(1700, 2001))

mean_dict = {
    'Year': year_list,
    'Mean PRATE': prate_mcrun_mean_list,
    'Mean PDSI': pdsi_mcrun_mean_list
}

mean_df = pd.DataFrame(mean_dict)
print(mean_df)
```

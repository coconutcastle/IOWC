```python
# FUNCTIONS

# name is a string, lat and lons are tuples indicating the start and stop point
# gets data for all mcruns
# both the prate and pdsi params go (time, mcrun, lat, lon)
def get_locale_pdsi_df(name, lat, lon, time_range):
    print(f'Converting PHYDA pdsi data for the locale {name}...')

    std_array = all_data['pdsi_sg'][time_range[0] - 1: time_range[1], lat[0]:lat[1], lon[0]:lon[1]]
    std_array = std_array.reshape(-1)

    prctl_5_array = all_data['pdsi_pc'][0, time_range[0] - 1: time_range[1], lat[0]:lat[1], lon[0]:lon[1]]
    prctl_5_array = prctl_5_array.reshape(-1)

    prctl_50_array = all_data['pdsi_pc'][1, time_range[0] - 1: time_range[1], lat[0]:lat[1], lon[0]:lon[1]]
    prctl_50_array = prctl_50_array.reshape(-1)

    prctl_95_array = all_data['pdsi_pc'][2, time_range[0] - 1: time_range[1], lat[0]:lat[1], lon[0]:lon[1]]
    prctl_95_array = prctl_95_array.reshape(-1)

    mean_array = all_data['pdsi_mn'][time_range[0] - 1: time_range[1], lat[0]:lat[1], lon[0]:lon[1]]
    mean_array = mean_array.reshape(-1)

    pdsi_dict = {
        'Year': list(range(time_range[0], time_range[1] + 1)),
        'PDSI 1 STD': std_array,
        'PDSI 5th PRCTL': prctl_5_array,
        'PDSI 50th PRCTL': prctl_50_array,
        'PDSI 95th PRCTL': prctl_95_array,
        'PDSI Mean': mean_array
    }

    pdsi_df = pd.DataFrame(pdsi_dict)

    print('Finished converting PHYDA data. Resultant dataframe is:')
    print(pdsi_df)

    return pdsi_df


# get pdsi dfs from lists containing data (assume all the same length)
def get_pdsi_dfs_from_locale_list(name_list, lat_list, lon_list, time_range=(1700, 2000)):

    locale_dfs = []

    for locale in range(len(name_list)):
        locale_pdsi = get_locale_pdsi_df(name_list[locale], lat_list[locale], lon_list[locale], time_range)
        locale_dfs.append(locale_pdsi)

    return locale_dfs
```

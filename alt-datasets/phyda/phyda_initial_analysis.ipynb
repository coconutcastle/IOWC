```python
for place in range(3):

    print('\n' + f'{place_names[place]}========================' + '\n')

    place_mean_array = all_data['pdsi_mn'][1699:, lats[place][0]:lats[place][1], lons[place][0]:lons[place][1]]
    place_mean_array = place_mean_array.reshape(-1)
    place_std_array = all_data['pdsi_sg'][1699:, lats[place][0]:lats[place][1], lons[place][0]:lons[place][1]]
    place_std_array = place_std_array.reshape(-1)

    place_prctl_5_array = all_data['pdsi_pc'][0, 1699:, lats[place][0]:lats[place][1], lons[place][0]:lons[place][1]]
    place_prctl_5_array = place_prctl_5_array.reshape(-1)
    place_prctl_50_array = all_data['pdsi_pc'][1, 1699:, lats[place][0]:lats[place][1], lons[place][0]:lons[place][1]]
    place_prctl_50_array = place_prctl_50_array.reshape(-1)
    place_prctl_95_array = all_data['pdsi_pc'][2, 1699:, lats[place][0]:lats[place][1], lons[place][0]:lons[place][1]]
    place_prctl_95_array = place_prctl_95_array.reshape(-1)

    place_dict = {
        'Year': list(range(1700, 2001)),
        'PDSI Mean': place_mean_array,
        'PDSI 1 STD': place_std_array,
        'PDSI 5th PRCTL': place_prctl_5_array,
        'PDSI 50th PRCTL': place_prctl_50_array,
        'PDSI 95th PRCTL': place_prctl_95_array
    }

    place_df = pd.DataFrame(place_dict)

    filename_place = f'C:\\Users\\Cecile Dai\\Documents\\Professional\\McGill University\\IOWC\\Other Datasets\\Alternative_Datasets\\PHYDA\\alt_places_{place_names[place]}.csv'

    place_df.to_csv(filename_place, index=False)

    print(place_df)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from itertools import combinations
```

```python
# Insert a list of all things you want to correlate and this will spit out every possible combo
def all_analysis(list_names, list_list):
    indices = list(range(len(list_names)))      #get a list of indices
    combos = list(combinations(indices, 2))     #get all possible combinations of indices, list of tuples

    for combo in combos:
        corr = pearsonr(list_list[combo[0]], list_list[combo[1]])[0]
        print(f'{list_names[combo[0]]} x {list_names[combo[1]]}: {corr}')
```

```python
tabora_reanalysis_df = pd.read_csv('tabora_reanalysis_redo.csv')
tabora_phyda_df = pd.read_csv('phyda_tabora_all_pdsi.csv')
tabora_lmr_df = pd.read_csv('../LMR/tabora_files/prate_pdsi_mcrun_mean.csv')
```

```python
tabora_reanalysis_array = tabora_reanalysis_df['Annual'].loc[tabora_reanalysis_df['Year'] <= 2000]
tabora_reanalysis_array = tabora_reanalysis_array.to_numpy()
print(tabora_reanalysis_array)
print("\n==================================\n")
tabora_phyda_array = tabora_phyda_df['PDSI Mean'].loc[tabora_phyda_df['Year'] >= 1836]
tabora_phyda_array = tabora_phyda_array.to_numpy()
print(tabora_phyda_array)
print("\n==================================\n")
tabora_lmr_pdsi_array = tabora_lmr_df['Mean PDSI'].loc[tabora_lmr_df['Year'] >= 1836]
tabora_lmr_pdsi_array = tabora_lmr_pdsi_array.to_numpy()
print(tabora_lmr_pdsi_array)
print("\n==================================\n")
tabora_lmr_prate_array = tabora_lmr_df['Mean PRATE'].loc[tabora_lmr_df['Year'] >= 1836]
tabora_lmr_prate_array = tabora_lmr_prate_array.to_numpy()
print(tabora_lmr_prate_array)
```

```python
array_list = [tabora_reanalysis_array, tabora_phyda_array, tabora_lmr_pdsi_array, tabora_lmr_prate_array]
array_names = ['Reanalysis Annual Precipitation', 'PHYDA Mean PDSI', 'LMR Mean PDSI', 'LMR Mean Precipitation']
```

```python
all_analysis(array_names, array_list)
```
